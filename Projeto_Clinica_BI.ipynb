```python
import pandas as pd
import numpy as np
from faker import Faker
```

```python
# Inicializar o gerenciador de dados para o portugues do Brasil
fake = Faker('pt_BR')

# Definindo a quantidade de medicos
qtd_medicos = 12
```

```python
#1. Gerar Tabela de Medicos com nomes aleatórios
medicos = pd.DataFrame({
    'ID_Medico': range(1, qtd_medicos + 1),
    # Gerar uma lista de nomes (usando prefixos como Dr/Dra.)
    'Nome': [fake.name() for _ in range(qtd_medicos)],
    'Especialidade': np.random.choice(
        ['Cardiologia', 'Pediatria', 'Ortopedia', 'Dermatologia', 'Neurologia', 'Ginecologia'], 
        qtd_medicos
    )
})
```

```python
# 2. Gerar Tabela de Salas
qtd_salas = 8
salas = pd.DataFrame({
    'ID_Sala': range(101, 101 + qtd_salas),
    'Nome_Sala': [f'Consultório {i}' for i in range(1, qtd_salas + 1)]
})
```

```python
# 3. Gerar Tabela de Agendamentos (500 linhas)
np.random.seed(42)
datas = pd.date_range(start='2026-04-01', periods=30, freq='D') # 30 dias de dados

agendamentos = pd.DataFrame({
    'Data': np.random.choice(datas, 500),
    'ID_Medico': np.random.choice(medicos['ID_Medico'], 500),
    'ID_Sala': np.random.choice(salas['ID_Sala'], 500),
    'Status': np.random.choice(['Confirmado', 'Faltou', 'Cancelado'], 500, p=[0.7, 0.2, 0.1])
})
```

```python
# Exportar
with pd.ExcelWriter('dados_clinica.xlsx') as writer:
    agendamentos.to_excel(writer, sheet_name='Agendamentos', index=False)
    medicos.to_excel(writer, sheet_name='Medicos', index=False)
    salas.to_excel(writer, sheet_name='Salas', index=False)

print(f"Planilha criada com {qtd_medicos} médicos e {len(agendamentos)} agendamentos!")
```
